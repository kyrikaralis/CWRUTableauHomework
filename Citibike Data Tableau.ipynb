```python
#import dependencies
import pandas as pd
import numpy as np
import csv
```

```python
#use 2 csv files for this exercise & concatenate them.
#this one is from Jan 2019. 
Jan19 = pd.read_csv("citibike-tripdata.csv")
Jan19.head()
```

```python
#2nd CSV, from Jan 2018 Citi Bike.
Jan18 = pd.read_csv("201801-citibike-tripdata.csv")
Jan18.head()
```

```python
#merge them together
df1 = pd.DataFrame(Jan18)
df2 = pd.DataFrame(Jan19)

Jan18_19_df = pd.concat([df1, df2])

Jan18_19_df.head()
```

```python
#Drop all rows where there's "NA"
Jan18_19_df.dropna(how='all')
```

```python
#export to CSV
Jan18_19_df.to_csv("Jan18_19.csv", encoding='utf-8', index=False)
```
